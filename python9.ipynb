```python
obp=float(input("obp puanınız:"))
yks_puan=float(input("yks puanınız:"))
puan_hesap=(0.6*obp) + yks_puan
print("---yks puanınız---")
print(puan_hesap)
eksikpuan= 415 - puan_hesap
if puan_hesap >= 415:
    print("tebrrikler puanınız tuttu eczacılığı kazandınız")

else:
    
    print(f"üzgünüz eczacılığa.{eksikpuan} ile puanınız yetmedi")
    
print(" {} obp puanı, {} yks puanı ile, tam puanınız {} dir ".format(obp,yks_puan,puan_hesap))
    
```

```python
geçiş_sorgusu={"ahmet":455,"mahmut":435,"ali":461,"alp":378,"tuna":388}
print("---kampüs sınav sonuçlar---")
for i in geçiş_sorgusu:
    puan=geçiş_sorgusu[i]
    if puan >= 417:
       print(f"tebrikler {i}, {puan} ile kampüsümüze girmeye hak kazandın:")

    elif puan < 417:
       print(f"üzgünüz {i}, {puan} ile kampüsümüze puanınız yetmiyor:")  
```
