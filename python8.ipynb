```python
puanlar={
"asel":85,"yavuz":92,"ramiz":90,"ali":77,"elif":80}
print("--- öğrenci karne durumları ---")
for i in puanlar:
    puan=puanlar[i]
    if puan >= 85:
        print(f"tebrikler {i}, {puan} ile taktir belgesi almaya hak kazandın")
    elif puan < 85:
        print(f"tebrikler {i}, {puan} ile teşekkür belgesi almaya hak kazandın")
```

```python
ehliyet={"ali":18,"murat":19,"merve":17,}
print("---ehliyet sınavı giriş kontrolü---")
for i in ehliyet:
    yaş=ehliyet[i]
    if yaş >= 18:
       print(f"tebrikler {i}, yaşınız ,{yaş}, gereği ehliyet alabilirsiniz")
    elif yaş < 18:
       print(f"sayın {i}, ,{yaş}, yaşında olmanız sebebi ile kanunen ehliyet alamazsınız")
```

```python
print("---ehliyet sınavı kontrolleri---")
yaş=int(input("lütfen yaşınızı girin:"))
sağlık_durumu=input("sağlık durumu nasıl:")
if yaş >= 18 and sağlık_durumu == "iyi":
    print(f"tebrikler ehliyet sınavına girebilirsin")

else:
    print(f"üzgünüz ehliyet sınavı almak için şartları karşılamıyorsunuz")
```
